### Standardizing Inconsistencies

### Visualisation

```python
# Import Statements
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Visualisation
# Histogram for all numeric columns
num_cols = df.select_dtypes(include=np.number) # Select only numeric columns
num_cols.hist(figsize=(10, 8))
plt.tight_layout()
plt.show()
```

```python
# Scatter Plot (Petal Length vs Petal Width)
plt.figure(figsize=(6,4))
sns.scatterplot(data=df, x='Petal Length', y='Petal Width', hue='Species')
plt.title("Petal Length vs Petal Width")
sns.regplot(data=df, x='Petal Length', y='Petal Width', scatter=False, color='red')
plt.show()
```

```python
# Scatter Plot (Sepal Length vs Sepal Width)
plt.figure(figsize=(6,4))
sns.scatterplot(data=df, x='Sepal Length', y='Sepal Width', hue='Species')
plt.title("Sepal Length vs Sepal Width")
sns.regplot(data=df, x='Sepal Length', y='Sepal Width', scatter=False, color='red')
plt.show()
```

```python
# Box Plot
plt.figure(figsize=(6,4))
sns.boxplot(data=df, x='Species', y='Sepal Length')
plt.title("Sepal Length by Species")
plt.show()
```

```python
# Correlation matrix
corr_matrix = num_cols.corr()
plt.figure(figsize=(6,4))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```

```python
# RdBu makes positive blue and negative red
plt.figure(figsize=(6,4))
sns.heatmap(corr_matrix, annot=True, cmap="RdBu", center=0)
plt.title("Correlation Matrix (Blue = Positive, Red = Negative)")
plt.show()
```
